```python
import tkinter as tk
from tkinter import messagebox as msgbox

class Zenith:
    def mut_services(self):
        mutual_services = ["Lines of credit", "Investment Management and accounts", "Insurance"]
        return mutual_services

    def uni_services(self):
        unique_services = [
            "Retirement and Education accounts", "Loans and Mortgages", "Checking and Savings",
            "Multi-currency management services and products", "Foreign Currency Accounts",
            "Foreign Currency Credit Cards", "Transborder Advisory Services",
            "Liquidity Management Services", "Advisory Services"
        ]
        return unique_services

class retail_banking(Zenith):
    def uni_services(self):
        unique_services = [
            "Lines of credit", "Investment Management and accounts", "Insurance",
            "Retirement and Education accounts", "Loans and Mortgages", "Checking and Savings"
        ]
        return unique_services

class global_banking(Zenith):
    def uni_services(self):
        unique_services = [
            "Multi-currency management services and products", "Foreign Currency Accounts",
            "Foreign Currency Credit Cards", "Transborder Advisory Services",
            "Liquidity Management Services"
        ]
        return unique_services

class commercial_banking(Zenith):
    def uni_services(self):
        unique_services = ["Advisory Services"]
        return unique_services

def click():
    first_name = first_name_entry.get()
    last_name = last_name_entry.get()
    name = (first_name + " " + last_name).title()
    division = division_entry.get().lower()

    if division == "retail":
        retail = retail_banking()
        all_services = retail.uni_services()
        msgbox.showinfo("Division Services", f"{name}, the services offered by your division, {division.capitalize()} Banking are:\n" + "\n".join(all_services))
    elif division == "global":
        glo = global_banking()
        all_services = glo.uni_services()
        msgbox.showinfo("Division Services", f"{name}, the services offered by your division, {division.capitalize()} Banking are:\n" + "\n".join(all_services))
    elif division == "commercial":
        comm = commercial_banking()
        all_services = comm.uni_services()
        msgbox.showinfo("Division Services", f"{name}, the services offered by your division, {division.capitalize()} Banking are:\n" + "\n".join(all_services))
    else:
        msgbox.showerror("ERROR!", f"DIVISION '{division}' is not recognised")

# GUI setup
root = tk.Tk()
root.title("Employee Attendance and Task Tracker")
root.geometry("500x200")

# First name
first_name_label = tk.Label(root, text="First Name:")
first_name_label.pack()
first_name_entry = tk.Entry(root)
first_name_entry.pack()

# Last name
last_name_label = tk.Label(root, text="Last Name:")
last_name_label.pack()
last_name_entry = tk.Entry(root)
last_name_entry.pack()

# Division
division_label = tk.Label(root, text="Division:")
division_label.pack()
division_entry = tk.Entry(root)
division_entry.pack()

# Submit button
button = tk.Button(root, text="SUBMIT", command=click)
button.pack()
button.config(fg="blue", bg="pink")

# Start GUI loop
root.mainloop()
```
